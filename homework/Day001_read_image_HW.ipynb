思考一下我們前面有提到圖片是矩陣，但維度可能會不一樣
例如灰階圖只有兩個維度，RGB 彩圖則有 3 個維度

假如今天我們把 RGB 3 個維度拆開來看會有甚麼不同的效果呢？

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

img_path = 'Part01/lena.png'
# 載入圖片(opencv是以BGR形式存取圖片)
img_bgr = cv2.imread(img_path, cv2.IMREAD_COLOR)
img_bgr.shape
```

### 將RBG 3個維度拆開，各會是2維的矩陣，畫出來會是灰階圖

```python
# 為了要不斷顯示圖片，所以使用一個迴圈
while True:
    cv2.imshow('BGR', img_bgr)
    cv2.imshow('B', img_bgr[:,:,0])
    cv2.imshow('G', img_bgr[:,:,1])
    cv2.imshow('R', img_bgr[:,:,2])
    # 直到按下 ESC 鍵才會自動關閉視窗結束程式
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
```

```python
# 將BGR轉為RGB
img_rgb = img_bgr[:,:,::-1]

plt.figure(figsize=(12,9))

plt.subplot(2,3,1)
plt.imshow(img_rgb)
plt.title('RGB')

plt.subplot(2,3,4)
plt.title('B')
plt.imshow(img_rgb[:,:,2], cmap ='gray')

plt.subplot(2,3,5)
plt.title('G')
plt.imshow(img_rgb[:,:,1], cmap ='gray')

plt.subplot(2,3,6)
plt.title('R')
plt.imshow(img_rgb[:,:,0], cmap ='gray')

plt.show()
```

### 若是只保留BGR其中一個維度之數值，其他維度改為0(保持3個維度)，繪圖結果如下:

```python
img_b, img_g, img_r = img_bgr.copy(), img_bgr.copy(), img_bgr.copy()
img_b[:,:,1:] = 0
img_g[:,:,[0,2]] = 0
img_r[:,:,0:2] = 0

while True:
    cv2.imshow('BGR', img_bgr)
    cv2.imshow('B', img_b)
    cv2.imshow('G', img_g)
    cv2.imshow('R', img_r)
    # 直到按下 ESC 鍵才會自動關閉視窗結束程式
    k = cv2.waitKey(0)
    if k == 27:
        cv2.destroyAllWindows()
        break
```

```python
plt.figure(figsize=(12,9))

plt.subplot(2,3,1)
plt.imshow(img_rgb)
plt.title('RGB')

plt.subplot(2,3,4)
plt.title('B')
plt.imshow(img_b[:,:,::-1])

plt.subplot(2,3,5)
plt.title('G')
plt.imshow(img_g[:,:,::-1])

plt.subplot(2,3,6)
plt.title('R')
plt.imshow(img_r[:,:,::-1])

plt.show()
```
